```python
pd_data_base=pd_raw.rename(columns={'Country/Region':'Country','Province/State':'State'})
pd_data_base['State']=pd_data_base['State'].fillna('No')
pd_data_base.head()
```

```python
pd_data_base = pd_data_base.drop(['Lat','Long'],axis=1).set_index(['State','Country']).T
pd_data_base.head()
```

```python
pd_relation_model=pd_data_base.stack(level=[0,1]).reset_index().rename(columns={'level_0':'date',0:'Confirmed'})
pd_relation_model
```

```python
pd_relation_model.dtypes
```

```python
pd_relation_model['date']=pd_relation_model.date.astype('datetime64[ns]')
pd_relation_model.dtypes
```

```python
pd_relation_model.to_csv('..\data\processed\COVID_relational_confirmed.csv',index=False)
```

# Analyzing data for Country US as it is in a new file due to the new structure of John Hopkins

```python
data_path='../data/raw/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'
pd_raw_US=pd.read_csv(data_path)
pd_raw_US.head()
```

```python
#Dropping the unnecessary data
pd_raw_US=pd_raw_US.drop(['UID', 'iso2', 'iso3', 'code3', 'Country_Region','FIPS', 'Admin2',  'Lat', 'Long_', 'Combined_Key'],axis=1)
pd_data_base_US=pd_raw_US.rename(columns={'Province_State':'State'}).copy()
```

```python
pd_relation_model_US=pd_data_base_US.set_index(['State']).T
pd_relation_model_US=pd_relation_model_US.stack().reset_index().rename(columns={'level_0':'date',0:'Confirmed'})
pd_relation_model_US['Country']='US'
pd_relation_model_US['date']=[datetime.strptime( each,"%m/%d/%y") for each in pd_relation_model_US.date]
```

```python
# Swap country and confirmed  to concatenate with previous columns
columns_titles = ["date","State","Country","Confirmed"]
pd_relation_model_US=pd_relation_model_US.reindex(columns=columns_titles)
pd_relation_model_US.head()
```

```python
pd_relation_model_all=pd_relation_model[pd_relation_model['Country']!='US'].reset_index(drop=True)
```

```python
pd_relation_model_all=pd.concat([pd_relation_model_all,pd_relation_model_US],ignore_index=True)
```

```python
pd_relation_model_all.to_csv('../data/processed/20220527_COVID_relational_confirmed.csv',index=False)
```
